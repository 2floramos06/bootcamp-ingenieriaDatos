Aqui se verifica si sirve la relacion entre emails, de students, customers y leads

```python
#emails en minusculas y sin espacios
emails_students = set(df_students['email'].dropna().str.lower().str.strip())
emails_customers = set(df_customers['email'].dropna().str.lower().str.strip())
emails_leads = set(df_leads['email'].dropna().str.lower().str.strip())

overlap_uni_billing = emails_students.intersection(emails_customers)
overlap_crm_billing = emails_leads.intersection(emails_customers)

print(f"Total emails únicos en University: {len(emails_students)}")
print(f"Total emails únicos en Billing: {len(emails_customers)}")
print(f"Total emails únicos en CRM: {len(emails_leads)}")
print("-" * 30)
print(f"Coincidencias reales University <-> Billing: {len(overlap_uni_billing)}")
print(f"Coincidencias reales CRM <-> Billing: {len(overlap_crm_billing)}")
```

Se puede evidenciar que solo existe una coincidencia por lo tanto la relacion no sirve. Asi que ahora se usa el atriuto external_ref de la entidad Customers, para comprobar si es igual a student_id

```python
df_cruce = pd.merge(
    df_customers,
    df_students,
    left_on='external_ref',
    right_on='student_id',
    suffixes=('_billing', '_uni'),
    how='inner'
)

print(f"Total de registros cruzados por ID: {len(df_cruce)}\n")

df_cruce['email_coincide'] = df_cruce['email_billing'] == df_cruce['email_uni']

print("¿Coinciden los emails para el mismo ID de estudiante?")
print(df_cruce['email_coincide'].value_counts())
print("\n" + "-"*50 + "\n")

columnas_ver = [
    'external_ref',
    'first_name_billing', 'first_name_uni',
    'email_billing', 'email_uni'
]
print("Muestra comparativa de los datos cruzados:")
print(df_cruce[columnas_ver].head())
```

Se puede evidenciar que se cruzaron los datos y si bien comparten el mismo id (ej. STU-0000001) no son las mismas personas

```python
tiene_contacto = df_activities['contact_id'].notna()
tiene_oportunidad = df_activities['opportunity_id'].notna()

ambos = (tiene_contacto & tiene_oportunidad).sum()

solo_contacto = (tiene_contacto & ~tiene_oportunidad).sum()

solo_oportunidad = (~tiene_contacto & tiene_oportunidad).sum()

ninguno = (~tiene_contacto & ~tiene_oportunidad).sum()

total = len(df_activities)

print(f"Total de registros en 'activities': {total}")
print("-" * 50)
print(f"1. Relación con AMBOS (Contacto y Oportunidad): {ambos} registros")
print(f"2. Relación SOLO con Contacto:                {solo_contacto} registros")
print(f"3. Relación SOLO con Oportunidad:             {solo_oportunidad} registros")
print(f"4. Relación con NINGUNO:          {ninguno} registros")
print("=" * 50)

suma_escenarios = ambos + solo_contacto + solo_oportunidad + ninguno
```
